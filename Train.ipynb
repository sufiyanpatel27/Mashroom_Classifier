```python
import pandas as pd
```

```python
df = pd.read_csv("data/mushrooms.csv")
```

```python
df.head()
```

```python
all_columns = df.columns.to_list()
for i in all_columns:
    nans = df[i].isnull().sum()
    if nans > 0:
        print(i, nans)
    else:
        print("no Nan values")
        break
```

#### Run this cell for Nan values

```python
for i in all_columns:
    nans = df[i].isnull().sum()
    if nans > 0:
        if str(type(df[i][0])) == "<class 'numpy.float64'>" or str(type(df[i][0])) == "<class 'int'>" :
            df[i] = df[i].fillna(df[i].mean())
        else:
            print('there are columns which have nan values in string format')
            print('deleteing those rows')
            df = df[df[i].notna()]
```

```python
from sklearn.preprocessing import LabelEncoder

for i in all_columns:
    if str(type(df[i][0])) == "<class 'str'>":
        encoder = LabelEncoder()
        df[i] = encoder.fit_transform(df[i])
```

```python
df.head()
```

```python
labels = df['class']
df.drop(['class'], axis = 1, inplace = True)
all_columns = df.columns.to_list()
```

```python
from sklearn.preprocessing import MinMaxScaler
for i in all_columns:
    scaler = MinMaxScaler()
    df[i] = scaler.fit_transform(df[[i]])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, labels,test_size=.2)
```

```python
from lazypredict.Supervised import LazyClassifier

clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)

print(models)
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf = AdaBoostClassifier(n_estimators=100, random_state=0)
clf.fit(X_train, y_train)
```

```python
print("train accuracy = ", int(clf.score(X_train, y_train)*100),"%")
```

```python
print("total cases = ", len(X_test))
right = 0
wrong = 0
y_pred = clf.predict(X_test)
for i in range(len(X_test)):
    if (y_pred[i] == y_test.values[i]):
        right += 1
    else:
        wrong += 1
print("right = ", right)
print("wrong = ", wrong)
print("test accuracy = ", int((len(X_test)/right)*100),"%")
```
